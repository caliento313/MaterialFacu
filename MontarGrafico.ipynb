```python
import matplotlib.pyplot as plt

meses=['janeiro','Fevereiro','Março','Abril','Maio']
vendas= [120,90,80,150,200]
plt.bar(meses,vendas)
plt.xlabel('Meses')
plt.ylabel('Vendas')
plt.title('Vendas por mês')
plt.show()
```
